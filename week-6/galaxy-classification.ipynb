### Galaxy classification
The dataset consists of galaxies identified as part of the Sloan Digital Sky Survey. For this task, the galaxy classes are limited to three types: spirals, ellipticals and mergers.



The galaxy catalogue is a sample of galaxies where at least 20 human classifiers (such as yourself) have come to a consensus on the galaxy type. Examples of spiral and elliptical galaxies were selected where there was a unanimous classification. Due to low sample numbers, we included merger examples where at least 80% of human classifiers selected the merger class.

The features used are:
- **colour indices**: colours (u-g, g-r, r-i, and i-z). Studies of galaxy evolution tell us that spiral galaxies have younger star populations and therefore are 'bluer' (brighter at lower wavelengths). Elliptical galaxies have an older star population and are brighter at higher wavelengths ('redder').
- **eccentricity**: approximates the shape of the galaxy by fitting an ellipse to its profile. Eccentricity is the ratio of the two axis (semi-major and semi-minor). The De Vaucouleurs model was used to attain these two axis. To simplify our experiments, we will use the median eccentricity across the 5 filters.
- **adaptive moments**: also describe the shape of a galaxy. They are used in image analysis to detect similar objects at different sizes and orientations. We use the fourth moment here for each band.
- **concentrations**: is similar to the luminosity profile of the galaxy, which measures what proportion of a galaxy's total light is emitted within what radius. A simplified way to represent this is to take the ratio of the radii containing 50% and 90% of the Petrosian flux. The Petrosian method allows us to compare the radial profiles of galaxies at different distances. If you are interested, you can read more [here](http://spiff.rit.edu/classes/phys443/lectures/gal_1/petro/petro.html) on the need for Petrosian approach.

These features are provided as part of the SDSS catalogue.

```python
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import cross_val_predict
from sklearn.ensemble import RandomForestClassifier
```

```python
# function that generates target and feature vectors
def generate_features_targets(data):
    # target
    y = data['class']
    # features
    X = np.empty(shape=(len(data), 13))
    X[:, 0] = data['u-g']
    X[:, 1] = data['g-r']
    X[:, 2] = data['r-i']
    X[:, 3] = data['i-z']
    X[:, 4] = data['ecc']
    X[:, 5] = data['m4_u']
    X[:, 6] = data['m4_g']
    X[:, 7] = data['m4_r']
    X[:, 8] = data['m4_i']
    X[:, 9] = data['m4_z']
    X[:, 10] = data['petroR50_u']/data['petroR90_u']
    X[:, 11] = data['petroR50_r']/data['petroR90_r']
    X[:, 12] = data['petroR50_z']/data['petroR90_z']

    return X, y
# function that visualizes the confusion matrix
def plot_confusion_matrix(actual_classes : np.array, predicted_classes : np.array, sorted_labels : list):

    matrix = confusion_matrix(actual_classes, predicted_classes, labels=sorted_labels)
    
    plt.figure(figsize=(12.8,6))
    sns.heatmap(matrix, annot=True, xticklabels=sorted_labels, yticklabels=sorted_labels, cmap="Blues", fmt="g")
    plt.xlabel('Predicted'); plt.ylabel('Actual'); plt.title('Confusion Matrix')

    plt.show()
```

```python
# import data
data = np.load('galaxy_catalogue.npy')

# generate the features and targets
X, y = generate_features_targets(data)

# instantiate a random forest classifier
n_estimators = 50 # number of individual decision trees
rfc = RandomForestClassifier(n_estimators=n_estimators)
  
# get predictions using 10-fold cross validation with cross_val_predict
y_pred = cross_val_predict(rfc, X, y, cv=10)

# calculate the model score using your function
accuracy = accuracy_score(y_pred, y)
print("Accuracy score:", accuracy)

# plot confusion matrix
plt.figure()
plot_confusion_matrix(y, y_pred, ["merger", "elliptical", "spiral"])
plt.show()
```
